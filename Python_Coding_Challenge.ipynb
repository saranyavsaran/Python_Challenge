# 2.Project Title: Secret Message Decoder
Scenario:

You’re building a Secret Message Decoder that helps a team decode hidden messages sent in plain text. The messages have mixed characters, spaces, and random numbers inserted to confuse readers. Your program should clean, transform, and analyze the text — without using in, if, or for.

Tasks:

Ask the user to input a secret message.

-Use len() and string methods (like .isalpha() ,lower(),upper()).

Convert the entire message to uppercase.

-Replace all spaces with underscores.

-Show only every 3rd character of the message.

-Use slicing: message[::3]

-Reverse the message and display it.

-Display the cleaned and transformed result.


```python
secret_message = input("Enter the Message")
length = len(secret_message)
change_case = secret_message.upper()
replaceing = secret_message.replace(" ", "_")
every_third_char = replaceing[::3]
reversed_message = replaceing[::-1]
print("Cleaned and Uppercase:", replaceing)
print("Length",length)
print("Every 3rd character:", every_third_char)
print("Reversed message:", reversed_message)


```

**4.Scenario**:

 You have been assigned to build a contact management system where employees can store and retrieve contact details efficiently.

**Tasks**:
Create a 2D list where each row holds a contact’s name, phone number, and email address.

Allow the user to add a new contact.

Display all contacts sorted alphabetically by name.

Allow the user to search for contacts by name. The search should support partial matches.

After sorting, display only the first 3 contacts in the list

```python
contacts = [
    ["Alice", "9876543210", "alice@example.com"],
    ["Bob", "9123456789", "bob@example.com"],
    ["Charlie", "9998887777", "charlie@example.com"],
    ["David", "9112233445", "david@example.com"],
    ["Eve", "9001122334", "eve@example.com"]
]
# print(contacts)
name  = input("Enter name: ")
phone = input("Enter phone: ")
email = input("Enter email: ")
contacts.append([name, phone, email])
contacts = sorted(contacts, key=str)
search_name = input("Enter the contact to search:")
result = [c for c in contacts if search_name in c[0].lower()]
print(result[:3])
```
